<hr style="border: 1px solid lightblue; width: 50%;">

<p style="font-family: Arial, sans-serif; color: #f0f0f0; font-size: 15px;">


<a id="introduction"></a>


<!-- Body Content -->
<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.6;">
Gestational Diabetes Mellitus (GDM) is one of the most common complications of pregnancy worldwide. It occurs when blood glucose levels rise during pregnancy due to hormonal and metabolic changes, creating risks for both the mother and the baby. If not managed, GDM can lead to complications such as preeclampsia, high birth weight, premature delivery, and an increased likelihood of developing type 2 diabetes later in life.
</p>

<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.6;">
Early detection and risk prediction of GDM are essential. They allow timely interventions through lifestyle changes, dietary management, and medical treatments, ultimately improving health outcomes for mothers and newborns. Since GDM often presents without obvious symptoms, many pregnant women may be unaware of their condition until complications arise, making proactive prediction models critical.
</p>

<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.6;">
The ability to predict GDM using machine learning offers healthcare providers evidence-based tools to recognize at-risk patients early, personalize care plans, and reduce adverse outcomes. By analyzing clinical and biometric data, machine learning models can uncover hidden risk factors and patterns that traditional screening methods may overlook.
</p>

<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.6;">
This study utilizes a curated GDM dataset to demonstrate the application of machine learning in healthcare. Through data preprocessing, feature analysis, model development, and evaluation of various classifiers, the project aims to highlight how predictive analytics can enhance maternal care and contribute to proactive healthcare solutions.
</p>

  <h3 style="color: lightblue; font-family: 'Trebuchet MS', sans-serif;">1.3 Algorithms </h3>
</center>

<!-- Divider -->
<hr style="border: 1px solid lightblue; width: 50%;">

<!-- Body Content -->
<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.6;">
To address the binary classification problem of predicting Positive (GDM) versus Negative (Non-GDM) cases, this project will apply several supervised machine learning algorithms. These algorithms have been carefully selected based on their proven effectiveness in medical datasets and their ability to handle clinical, demographic, and biometric features.
</p>

<!-- Bullet List -->
<ul style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.8;">
  <li><strong>Logistic Regression</strong>: A well-established linear model widely used for binary classification. It provides interpretable coefficients, making it useful as a baseline model and for clinical explainability.</li>
  
  <li><strong>Support Vector Machine (SVM)</strong>: Performs well with smaller datasets and high-dimensional data. By using kernels (such as RBF), it can model complex nonlinear decision boundaries effectively.</li>
  
  <li><strong>K-Nearest Neighbors (KNN)</strong>: A simple, instance-based algorithm that classifies patients based on similarity to their nearest neighbors in the feature space. It works well when decision boundaries are highly local.</li>
  
  <li><strong>Random Forest</strong>: An ensemble method that aggregates multiple decision trees to improve classification accuracy. It also provides valuable feature importance measures, helping identify significant predictors of GDM risk.</li>
  
  <li><strong>Naive Bayes</strong>: A probabilistic classifier that assumes independence among features. It is computationally efficient and performs well on smaller datasets, making it a useful benchmark algorithm.</li>
  
  <li><strong>XGBoost (Extreme Gradient Boosting)</strong>: A state-of-the-art gradient boosting algorithm that often outperforms traditional methods on structured/tabular medical data. It handles missing values, captures complex feature interactions, and manages class imbalance effectively, making it highly suitable for clinical prediction tasks.</li>
</ul>

<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.6;">
These models will be trained, evaluated, and compared to determine the most effective classifier for predicting the risk of GDM in the provided dataset. Special consideration will be given to class imbalance, using techniques such as weighted models or oversampling to ensure fair evaluation.
</p>




<!-- Subsection Title -->
<center> 
  <h4 style="color: #87cefa; font-family: 'Trebuchet MS', sans-serif;">1.3.1 Why These Algorithms for GDM?</h4>
</center>

<!-- Subsection Divider -->
<hr style="border: 0.8px solid #87cefa; width: 30%;">

<!-- Body Content -->
<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 13.5px; line-height: 1.6;">
The prediction of Gestational Diabetes Mellitus (GDM) depends on analyzing clinical and demographic features such as maternal age, BMI, weight, height, fasting glucose, and oral glucose tolerance test (OGTT) values. These features interact in nonlinear and complex ways, which makes selecting the right algorithms essential. The chosen models balance interpretability with predictive power, making them well-suited for medical decision support in maternal care.
</p>

<ul style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 13.5px; line-height: 1.8;">
  <li><strong>Support Vector Machine (SVM)</strong>: GDM risk factors such as BMI and OGTT readings often cannot be separated linearly. SVM, particularly with the RBF kernel, can handle nonlinear boundaries and high-dimensional feature spaces, making it well-suited for clinical data where relationships are subtle but important.</li>

  <li><strong>K-Nearest Neighbors (KNN)</strong>: Pregnant women with similar characteristics (e.g., similar OGTT patterns or BMI values) may share the same diagnosis. KNN leverages this idea by classifying patients based on similarity to their nearest neighbors, making it effective in identifying local clusters of at-risk women.</li>

  <li><strong>Random Forest</strong>: GDM is influenced by many interacting features (e.g., age, BMI, glucose levels). Random Forest handles nonlinear interactions, manages noise, and provides feature importance rankings, helping identify which factors contribute most to GDM risk prediction.</li>

  <li><strong>Logistic Regression</strong>: Logistic Regression remains a fundamental and interpretable model in binary classification. In GDM prediction, its coefficients clearly show how risk factors like maternal age or fasting glucose increase the probability of developing GDM, providing transparency for clinical decisions.</li>

  <li><strong>Naive Bayes</strong>: Despite its independence assumption, Naive Bayes is simple, fast, and effective on smaller datasets. It serves as a benchmark algorithm and offers a probabilistic approach that is useful for comparison with more complex classifiers.</li>

  <li><strong>XGBoost (Extreme Gradient Boosting)</strong>: GDM prediction benefits from advanced ensemble methods. XGBoost captures complex interactions between features, handles missing data efficiently, and mitigates class imbalance, often outperforming traditional classifiers in structured medical datasets.</li>
</ul>

<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 13.5px; line-height: 1.6;">
Together, these algorithms provide a strong toolkit for GDM prediction. Models like Logistic Regression and Random Forest prioritize interpretability for healthcare practitioners, while SVM and XGBoost emphasize predictive power. This balance ensures the project can both achieve high performance and provide clinically relevant insights.
</p>

Recent advances in machine learning have enabled earlier prediction and classification of Gestational Diabetes Mellitus (GDM) by leveraging maternal demographic and clinical datasets. In one study, Jayakumar et al. compared the performances of Decision Trees, Naive Bayes, Random Forest, and SVM, where Random Forest consistently achieved accuracies above 90% in multiple clinical datasets, reinforcing the utility of ensemble learning in medical prediction tasks <strong>[1]</strong>.

Finally, Chauhan and Patel addressed dataset imbalance using SMOTE with Random Forest and XGBoost, significantly improving recall for minority cases, demonstrating the importance of class rebalancing techniques in medical datasets <strong>[5]</strong>.

<center> 
  <h3 style="color: lightblue; font-family: 'Trebuchet MS', sans-serif;">3.2 Feature Understanding</h3>
</center>

<hr style="border: 1px solid lightblue; width: 50%;">

<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.6;">
This dataset consists of clinical, demographic, and diagnostic features that are associated with gestational diabetes mellitus (GDM) and non-gestational diabetes (NGDM) cases. Below is a description of each feature along with its potential relevance to maternal and neonatal health outcomes:
</p>

<ul style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.8;">

  <li><strong>Patients</strong>: Unique identifier assigned to each patient (e.g., PT1, PT2). Used only for indexing and not for model training.</li>

  <li><strong>Type of Treatment</strong>: (GDM sheet only) Indicates the treatment received by the patient (e.g., Metformin, Insulin, Diet). Essential for analyzing the effectiveness of different therapeutic approaches.</li>

  <li><strong>Age (Years)</strong>: Maternal age at diagnosis. Higher maternal age is a known risk factor for gestational diabetes and pregnancy complications.</li>

  <li><strong>Nationality</strong>: Country of origin or citizenship of the patient. Useful for understanding demographic and ethnic variations in outcomes.</li>

  <li><strong>BMI at Baseline (Kg/m²)</strong>: Body Mass Index recorded at baseline. Elevated BMI is strongly correlated with increased GDM risk.</li>

  <li><strong>Gestational Age at Diagnosis (Months)</strong>: The month of pregnancy when GDM was diagnosed. Early or late diagnosis can influence treatment outcomes.</li>

  <li><strong>Fasting Blood Glucose</strong>: Blood glucose level after fasting. A key diagnostic indicator for diabetes.</li>

  <li><strong>1 hour after 75 g of glucose</strong>: Plasma glucose concentration 1 hour after oral glucose tolerance test (OGTT). Used as part of standard GDM screening.</li>

  <li><strong>2 hours after 75 g of glucose</strong>: Plasma glucose concentration 2 hours after OGTT. Helps confirm impaired glucose tolerance.</li>

  <li><strong>HbA1c Levels at Diagnosis</strong>: Glycated hemoglobin level recorded at diagnosis (mainly for patients on insulin). Reflects long-term blood glucose control.</li>

  <li><strong>HbA1c Levels at Delivery</strong>: Glycated hemoglobin at the time of delivery. Provides insight into glycemic control throughout pregnancy.</li>

  <li><strong>Weight Gain During Pregnancy (Kg)</strong>: Total maternal weight gain during pregnancy. Both excessive and insufficient gain may affect maternal and neonatal health.</li>

  <li><strong>Hypertensive Disorders (Yes/No)</strong>: Indicates the presence of hypertension-related conditions during pregnancy. Strongly associated with maternal and fetal complications.</li>

  <li><strong>Mode of Delivery</strong>: (NGDM sheet only) Refers to childbirth method (e.g., Cesarean or Natural). Allows for comparative analysis between GDM and NGDM groups.</li>

  <li><strong>Weight (Kg)</strong>: Maternal weight measurement. Complements BMI and weight gain analysis.</li>

  <li><strong>Height (cm)</strong>: Maternal height in centimeters. Used alongside weight to calculate BMI.</li>

  <li><strong>Pulse/Heart Rate (bpm)</strong>: Maternal heart rate in beats per minute. Provides insights into cardiovascular health.</li>

  <li><strong>BP Systolic (mmHg)</strong>: Systolic blood pressure measurement. Elevated values are a risk factor for hypertensive disorders.</li>

  <li><strong>BP Diastolic (mmHg)</strong>: Diastolic blood pressure measurement. Together with systolic pressure, it helps diagnose hypertension.</li>

  <li><strong>Socioeconomic Status</strong>: Represents the patient’s socioeconomic background. May indirectly influence lifestyle, healthcare access, and outcomes.</li>

</ul>

<p style="font-family: Arial, sans-serif; color: #dcdcdc; font-size: 14px; line-height: 1.6;">
Each of these features provides valuable demographic or clinical information that contributes to assessing maternal health, understanding risk factors, and predicting gestational diabetes outcomes. They will serve as inputs for the machine learning models built for this study.
</p>

```python
import numpy as np, pandas as pd, xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, accuracy_score, precision_recall_fscore_support

def _slice(X, idx):
    if isinstance(X, pd.DataFrame): return X.iloc[idx].values
    if isinstance(X, pd.Series):    return X.iloc[idx].values
    return X[idx]

def eval_block(name, X, y, params, split_seed=42, leak=False, smote_before=False, thresholds=(0.50,0.55,0.60)):
    y = np.asarray(y).astype(int)
    if leak:
        imp = SimpleImputer(strategy="median")
        X_imp = imp.fit_transform(_slice(X, np.arange(len(y))))
        tr, te = train_test_split(np.arange(len(y)), test_size=0.2, stratify=y, random_state=split_seed)
        X_tr, X_te = X_imp[tr], X_imp[te]
        y_tr, y_te = y[tr], y[te]
        if smote_before:
            try:
                from imblearn.over_sampling import SMOTE
                X_tr, y_tr = SMOTE(random_state=split_seed).fit_resample(X_tr, y_tr)
            except Exception:
                pass
    else:
        tr, te = train_test_split(np.arange(len(y)), test_size=0.2, stratify=y, random_state=split_seed)
        X_tr_raw, X_te_raw = _slice(X, tr), _slice(X, te)
        y_tr, y_te = y[tr], y[te]
        imp = SimpleImputer(strategy="median")
        X_tr = imp.fit_transform(X_tr_raw)
        X_te  = imp.transform(X_te_raw)

    clf = xgb.XGBClassifier(objective="binary:logistic", eval_metric="logloss", random_state=42, tree_method="hist", **params)
    clf.fit(X_tr, y_tr)
    proba = clf.predict_proba(X_te)[:,1]

    print(f"\n=== {name} ===")
    for thr in thresholds:
        y_pred = (proba >= thr).astype(int)
        acc = accuracy_score(y_te, y_pred)
        p,r,f1,_ = precision_recall_fscore_support(y_te, y_pred, average="micro", zero_division=0)
        print(f"\nThreshold: {thr:.2f}")
        print(classification_report(y_te, y_pred, digits=2))
        print(f"Micro: Precision {p:.2f} | Recall {r:.2f} | F1 {f1:.2f} | Accuracy {acc:.2f}")

det_params = dict(n_estimators=300, learning_rate=0.05, max_depth=6, min_child_weight=7, gamma=0.2,
                  subsample=1.0, colsample_bytree=1.0, reg_alpha=0.0, reg_lambda=2.0)
tuned_subsample_params = dict(det_params, subsample=0.7, colsample_bytree=0.7)

eval_block("Honest_Deterministic", X, y, det_params)
eval_block("WithSubsample_Tuned", X, y, tuned_subsample_params)
eval_block("LeakyCheck_FullImpute_SMOTEBeforeSplit", X, y, det_params, leak=True, smote_before=True)
```

```python
from sklearn.metrics import auc



def plot_roc_curve(model, X_test, y_test, model_name):
    try:
        # Check if the model supports predict_proba
        if hasattr(model, "predict_proba"):
            y_probs = model.predict_proba(X_test)[:, 1]  # Select probability for the positive class
        else:
            # If no probability estimates, use decision_function (for models like SVM)
            y_probs = model.decision_function(X_test)

        # Compute ROC curve and AUC
        fpr, tpr, _ = roc_curve(y_test, y_probs)
        roc_auc = auc(fpr, tpr)

        # Plot ROC Curve
        plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

    except Exception as e:
        print(f"Could not generate ROC for {model_name}: {e}")


plt.figure(figsize=(8, 6))

# Plot for Logistic Regression
plot_roc_curve(baseline_model, X_test, y_test, "Logistic Regression")

# Plot for SVM (Only if using probability=True)
if hasattr(svm_model, "predict_proba"):
    plot_roc_curve(svm_model, X_test, y_test, "SVM")

# Plot for KNN
plot_roc_curve(knn_model, X_test, y_test, "KNN")

# Plot for Random Forest
plot_roc_curve(rf_model, X_test, y_test, "Random Forest")

# Plot for Naive Bayes
plot_roc_curve(nb_model, X_test, y_test, "Naive Bayes")

# Reference diagonal line (random guessing)
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")

# Customize plot
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves for Classification Models")
plt.legend()
plt.show()
```
